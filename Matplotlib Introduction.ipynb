## The basics
[matplotlib.pyplot](https://matplotlib.org/api/pyplot_api.html#module-matplotlib.pyplot) provides a plotting environment for Python. It's easy and fast to plot a range of visualizations. For example, let's plot the array `[1,2,3,4]`:

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.show()
```

### Adding titles, axis labels, and a legend
Let's redraw this plot but now with a title, axis labels, and a legend:

```python
x_vals = [1,2,3,4]
plt.plot(x_vals, label="An awesome line")
plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title("The title of the graph!")
plt.legend()
plt.show()
```

### Adding both x and y data

You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to the [plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you.

[plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) is a versatile command, and will take an arbitrary number of arguments. For example, to plot x versus y, you can issue the command:

```python
x_vals = [1,2,3,4]
y_vals = [1, 4, 9, 16]
plt.plot(x_vals, y_vals, label='Data!')
plt.legend()
plt.show() # You must call plt.show() to make graphics appear
```

Here's another example using NumPy:

```python
import numpy as np
import matplotlib.pyplot as plt # pyplot is mainly intended for interactive plots and programmatic plot generation

# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()  # You must call plt.show() to make graphics appear
```

### Setting axis properties
You can set the [xmin, xmax, ymin, ymax] of the plot using the [axis()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html) command. So, for example:

```python
import matplotlib.pyplot as plt
plt.plot(x_vals, y_vals)
plt.axis([0, 6, 0, 20]) # x axis from 0 to 6, y axis from 0 to 20
plt.show()
```

We can also set custom labels, including rotating the labels if they are too long.

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 6]
labels = ['A long label', 'A longer label', 'A much longer label', 'Label!']

plt.plot(x, y)
plt.xticks(x, labels, rotation=45) # You can specify a rotation for the tick labels in degrees or with keywords - try it with '45'
plt.margins(0.2) # Pad margins so that markers don't get clipped by the axes
plt.show()
```

### Resizing the plot
We can also resize the plot using `figsize` option in [figure()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html). This will be important later when we make multiple plots together.

```python
import random

num_rand_data_points = 17
plt.figure(figsize=(12, 5), facecolor='w', edgecolor='k') # figsize is width, height
x_vals = [random.randrange(1, 50, 1) for i in range(num_rand_data_points)]
plt.plot(x_vals, label="Random data")
plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title("{} random data points!".format(num_rand_data_points)) # programmatically include 7 in the title - try changing the number of points and see what happens
plt.legend()
plt.show()
```

### Plotting multiple values
You can plot multiple lines too:

```python
# setup the data
(x1_vals, y1_vals) = ([1,2,3,4], [1, 4, 9, 16])
(x2_vals, y2_vals) = ([0,1,1.5,7], [1,4,12,16])
(x3_vals, y3_vals) = ([0,1,3,5], [3,4.5,7,18])

# plot the data
plt.figure(figsize=(10, 5))
plt.plot(x1_vals, y1_vals)
plt.plot(x2_vals, y2_vals)
plt.plot(x3_vals, y3_vals)
plt.show()
```

An even more compact way of doing this:

```python
plt.figure(figsize=(10, 5))
plt.plot(x1_vals, y1_vals, x2_vals, y2_vals, x3_vals, y3_vals)
plt.show()
```

And again with NumPy:

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

## Controlling Line Properties
- https://matplotlib.org/stable/tutorials/introductory/pyplot.html#controlling-line-properties

Lines have many attributes that you can set: linewidth, dash style, marker, antialiasing, alpha, and more; see [matplotlib.lines.Line2D](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html). There are several ways to set line properties

```python
(x1_vals, y1_vals) = ([1,2,3,4], random.sample(range(0, 5), 4))
(x2_vals, y2_vals) = ([1,2,3,4], random.sample(range(0, 5), 4))
(x3_vals, y3_vals) = ([1,2,3,4], random.sample(range(0, 5), 4))

line1 = plt.plot(x1_vals, y1_vals)
line2 = plt.plot(x2_vals, y2_vals)
line3 = plt.plot(x3_vals, y3_vals)
plt.setp(line1, color='r', linewidth=2.0, marker='D', alpha=0.5) # red line with diamond markers
plt.setp(line2, color='b', linewidth=2.0, marker='X', alpha=0.5) # blue line with X markers
plt.show()
```

```python
lines = plt.plot(x1_vals, y1_vals, x2_vals, y2_vals, x3_vals, y3_vals) # do the same as above but more concise with our code
plt.setp(lines[0], color='r', linewidth=2.0, marker='D', alpha=0.5)
plt.setp(lines[1], color='b', linewidth=2.0, marker='X', alpha=0.5)
plt.show()
```

You can also set these properties using keyword args in the plot function:

```python
line1 = plt.plot(x1_vals, y1_vals, color='r', linewidth=2.0, marker='D', alpha=0.5)
line2 = plt.plot(x2_vals, y2_vals, color='b', linewidth=2.0, marker='X', alpha=0.5)
line3 = plt.plot(x3_vals, y3_vals)
plt.show()
```

## Plotting multiple charts

You can create multiple plots within the same figure by using [`subplot`](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplot.html)

**References:**
- https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplot.html
- https://matplotlib.org/devdocs/gallery/subplots_axes_and_figures/subplots_demo.html

```python
import numpy as np
import matplotlib.pyplot as plt

# subplot adds a subplot to the figure. To set the subplot location, pass
# in a three digit integer (< 10) where the first digit is the number of rows,
# the second the number of columns, and the third the index of the
# subplot. So, plt.subplot(211) refers to a two row plot with one
# column and selects the subplot index. This is equivalent to
# plt.subplot(2,1,1). See:
#   https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html
plt.subplot(211)
plt.plot(x1_vals, y1_vals)

plt.subplot(212)
plt.plot(x2_vals, y2_vals, 'r-') # the 'r-' colors the line red
plt.show()
```

```python
# We can also create a grid by looping through each plot
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4) # The command plt.subplots_adjust can be used to adjust the spacing between these plots
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

## Specifying Colors, Markers

### Colors
- https://matplotlib.org/stable/api/colors_api.html#matplotlib-colors
- https://matplotlib.org/stable/tutorials/colors/colors.html

In almost all places in matplotlib where a color can be specified by the user it can be provided as:

- an RGB or RGBA tuple of float values in [0, 1] (e.g., (0.1, 0.2, 0.5) or (0.1, 0.2, 0.5, 0.3))
- a hex RGB or RGBA string (e.g., '#0F0F0F' or '#0F0F0F0F')
- a string representation of a float value in [0, 1] inclusive for gray level (e.g., '0.5')
one of {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
- a X11/CSS4 color name
- a name from the xkcd color survey prefixed with 'xkcd:' (e.g., 'xkcd:sky blue')
- one of {'C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9'}
- one of {'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'} which are the Tableau Colors from the T10 categorical palette (which is the default color cycle).

For the basic built-in colors, you can use a single letter

- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white

- "v" triangle_down

## Creating dates

We often plot timeseries data as a function of time using data collected on specific dates. There is a useful Numpy function [datetime](https://numpy.org/doc/stable/reference/arrays.scalars.html#numpy.datetime64) to create an array of evenly spaced dates that can be used as x-axes values for plotting timeseries data:

```python
times = np.arange(np.datetime64('2001-01-02'),
                  np.datetime64('2002-02-03'), np.timedelta64(120, 'm')) # specify the start and end date as YYYY-MM-DD
#The arguments for timedelta64 are a number, to represent the number of units, and a date/time unit, such as (D)ay, (M)onth, (Y)ear, (h)ours, (m)inutes, or (s)econds
y = np.random.randn(len(times))

fig, ax = plt.subplots()
ax.plot(times, y)
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, '*', color='green');
```

## Histograms

A simple histogram can be a great first step in understanding a dataset. We can create them using the [plt.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) function. Below are some basic, minimum working examples:

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000) # create a simple histogram for random data points
plt.hist(data);
```

```python
# Plot the same data with more bins. Use edge colors to better identify bars
plt.hist(data, bins=30, alpha=0.5, color='steelblue', edgecolor='black');
```

```python
x1 = np.random.normal(0, 0.8, 1000) # create three random normal distributions
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, alpha=0.3, bins=40)
plt.hist(x2, alpha=0.3, bins=40)
plt.hist(x3, alpha=0.3, bins=40);
```

## Bar charts

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically.

The following is a simple example of the bar plot, which represents the number of students enrolled in different courses of an institute.


```python
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys()) # put the data into lists
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

plt.style.use('seaborn') # use the seaborn style - for all available styles, see here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

# creating the bar plot
plt.bar(courses, values,
        width = 0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
