# What are Jupyter Notebooks?

[Jupyter notebooks](https://jupyter.org/) are documents that can be viewed and executed inside any modern web browser. For this course, we will use the [Google Colaboratory (or 'Colab')](https://colab.research.google.com/) implementation of Jupyter. Since you're reading this notebook, you already know how to view a Jupyter notebook. The next step is to learn how to execute code that may be embedded in a Jupyter notebook.

To execute Python code in a notebook you will need access to a Python kernel. A kernel is simply a program that runs in the background, maintains workspace memory for variables and functions, and executes Python code. The kernel can be located on the same laptop as your web browser or located in an online cloud service (Google Colab). 

**Important Note Regarding Versions** There are two versions of Python in widespread use. Version 2.7 released in 2010, which was the last release of the 2.x series. Version 3.10 is the most recent release of the 3.x series which represents the future direction of language. It has taken years for the major scientific libraries to complete the transition from 2.x to 3.x, but it is now safe to recommend Python 3.x for widespread use. So for this course be sure to use version 3.x of the Python language (latest 3.10).
